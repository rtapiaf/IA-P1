```python
import pandas as pd
data = pd.read_csv("student-mat.csv")
data.head()
```

```python
X = pd.DataFrame(data, columns=["age","studytime","famrel","freetime","goout"])
X.head()
```

```python
y = pd.DataFrame(data, columns=["sex"])
y.head()
```

```python
import numpy as np
X1=np.array(X)
y1=np.array(y)
print(X1)
print(y1)
```

```python
from sklearn import tree
clasificador = tree.DecisionTreeClassifier()
clasificacion = clasificador.fit(X,y)
print(clasificacion)
```

```python
prediccion=clasificacion.predict([[16,2,3,2,2]])
print(prediccion)
```
