Seleccionamos el Dataset "Consumo de Alcohol por parte de los Estudiantes"


```python
import pandas as pd
data = pd.read_csv("student-mat.csv")
data.head()
```

Seleccionamos la columna a trabajar "Edad"

```python
lista = data["age"]
print(lista)
```

Hallamos las edades y la cantidad de estudiantes con estas edades

```python
import matplotlib.pyplot as plt
valores_x = data["age"].unique()
valores_y = data["age"].value_counts().tolist()
print(valores_x, valores_y)
```

Graficamos

```python
plt.bar(valores_x, valores_y)
plt.title("SepalLengthCm")
ax=plt.subplot()
#ax.set_xticks(valores_x)
ax.set_xticklabels(valores_x)
ax.set_xlabel("edad")
ax.set_ylabel("NroEst")
plt.show()
```

Seleccionamos el Dataset "World University Rankings"

```python
import pandas as pd
data = pd.read_csv("shanghaiData.csv")
data.head()
```

Seleccionamos la columna a trabajar "pub" que son el numero total articulos indexados y citados en publicaciones cientificas

```python
lista = data["pub"]
print(lista)
```

Hallamos los indices de las publicaciones citadas y la cantidad de estas

```python
import matplotlib.pyplot as plt
valores_x = data["pub"].unique()
valores_y = data["pub"].value_counts().tolist()
print(valores_x, valores_y)
```

Graficamos

```python
plt.bar(valores_x, valores_y)
plt.title("pub")
ax=plt.subplot()
#ax.set_xticks(valores_x)
ax.set_xticklabels(valores_x)
ax.set_xlabel("PubCitadas")
ax.set_ylabel("Cantidad")
plt.show()
```
